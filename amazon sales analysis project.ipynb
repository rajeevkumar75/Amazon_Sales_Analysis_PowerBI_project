# Amazon sales analysis:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('/content/drive/MyDrive/Project/Amazon sales analysis/amazon (1).csv')
df.head()
```

```python
#looking for null value:
plt.figure(figsize = (18,6))
sns.heatmap(df.isnull(), cmap='plasma', yticklabels=False, cbar=False)
plt.title('Missing Data By columns',fontsize=20)
plt.xticks(fontsize=15)
plt.show()
print(df.isnull().sum())
```

```python
df.info()
```

```python
df['category_new'] = df['category'].str.split().str[:1].str.join('|')
```

```python
df['product_name_new'] = df['product_name'].str.split().str[:9].str.join(' ')
```

```python
df.head()
```

##About Data:

```python
Total_sales = np.sum(df['discounted_price'])/1000000
print('About Data:- \n Total_sales: $',np.round(Total_sales,2),'M')

Avg_sales = np.average(df['discounted_price'])/1000
print(' Average Sales: $',np.round(Avg_sales,2),'K')

No_of_items = np.count_nonzero(df['product_id'])/1000
print(' Number of Items:',np.round(No_of_items,2),'K')

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
Avg_rating = np.average(df['rating'].dropna())
print(' Average Rating:', np.round(Avg_rating,1))
```

#**Total Sales:**

###Total sales by category:

```python
category_sales_sum = df.groupby('category_new')['discounted_price'].sum()
category_product_count = df.groupby('category_new')['product_name'].count()

df2 = pd.DataFrame({
    'Category': category_sales_sum.index,
    'Total Sales': category_sales_sum.values,
    'Number of Products': category_product_count.values
    })

fig = px.bar(
    df2,
    x='Total Sales',
    y='Category',
    title='Total Sales by Category',
    orientation='h',
    width=1500,
    height=800,
    hover_data={'Number of Products': True}
    )

fig.update_traces(marker_color='red')
fig.update_layout(
    yaxis=dict(
        automargin=True,
        categoryorder='total ascending'
    )
)
fig.show()
print(category_sales_sum.sort_values(ascending=False))
```

###Total Sales by Product Name:

```python
product_sales_sum = df.groupby('product_name_new')['discounted_price'].sum()
df1 = pd.DataFrame({
    'Product Name': product_sales_sum.index,
    'Total Sales': product_sales_sum.values
    })

fig = px.bar(df1,
             x='Total Sales',
             y='Product Name',
             title='Total Sales by Product',
             orientation='h',
             width=1500,
             height=800)

fig.update_traces(marker_color='red')
fig.update_layout(
    yaxis=dict(
        automargin=True,
        categoryorder='total ascending'
    ))

fig.show()
print(product_sales_sum.sort_values(ascending=False))
```

#**Average Sales:**

###Average Sales by category:

```python
category_sales_Avg = df.groupby('category_new')['discounted_price'].mean()
df3 = pd.DataFrame({
    'Category': category_sales_Avg.index,
    'Average Sales': category_sales_Avg.values
    })

fig = px.bar(df3,
             x='Average Sales',
             y='Category',
             title='Average Sales by Category',
             orientation='h',
             width=1500,
             height=800)

fig.update_traces(marker_color='red')
fig.update_layout(
    yaxis=dict(
        automargin=True,
        categoryorder='total ascending'
    ))

fig.show()
print(category_sales_Avg.sort_values(ascending=False))
```

###Average Sales by Product:

```python
product_sales_Avg = df.groupby('product_name_new')['discounted_price'].mean()
df4 = pd.DataFrame({
    'Product Name': product_sales_Avg.index,
    'Average Sales': product_sales_Avg.values
    })

fig = px.bar(df4,
             x='Average Sales',
             y='Product Name',
             title='Average Sales by Product',
             orientation='h',
             width=1500,
             height=800)

fig.update_traces(marker_color='red')

fig.update_layout(
    yaxis=dict(
        automargin=True,
        categoryorder='total ascending'
    ))

fig.show()
print(product_sales_Avg.sort_values(ascending=False))
```

#**Discount Analysis**
Analyze which categories frequently offer the highest discounts:

##Max discount by category:

```python
max_discount_by_category = df.groupby('category_new')['discount_percentage'].max().sort_values(ascending=False)

max_discount_df = pd.DataFrame({
    'Category': max_discount_by_category.index,
    'Maximum Discount (%)': max_discount_by_category.values*100
    })

fig_max_discount = px.bar(
    max_discount_df,
    x='Maximum Discount (%)',
    y='Category',
    title='Maximum Discount by Category',
    labels={'Category': 'Category', 'Maximum Discount (%)': 'Maximum Discount (%)'},
    color='Maximum Discount (%)',
    color_continuous_scale='Plasma',
    orientation='h'
    )

fig_max_discount.update_layout(
    xaxis=dict(
        title='Maximum Discount (%)',
        tickmode='linear',
        range=[0, max_discount_by_category.max() * 1.1],
        autorange=False
    ),
    yaxis=dict(
        title='Category',
        automargin=True
    ),
    height=800,
    width=1500,
    margin=dict(r=20, t=50, l=150, b=50))

fig_max_discount.show()

print(max_discount_by_category*100)
```

##Maximum Discount by Product:

```python
max_discount_by_product_name = df.groupby('product_name_new')['discount_percentage'].max().sort_values(ascending=False)
max_discount_df2 = pd.DataFrame({
    'Product Name': max_discount_by_product_name.index,
    'Maximum Discount (%)': max_discount_by_product_name.values*100
    })

fig_max_discount = px.bar(
    max_discount_df2,
    x='Maximum Discount (%)',
    y='Product Name',
    title='Maximum Discount percent by Product Name',
    labels={'Product Name': 'Product Name', 'Maximum Discount (%)': 'Maximum Discount (%)'},
    color='Maximum Discount (%)',
    color_continuous_scale='Plasma',
    orientation='h'
    )

fig_max_discount.update_layout(
    xaxis=dict(
        title='Maximum Discount (%)',
        tickmode='linear',
        range=[0, max_discount_by_product_name.max() * 1.1],
        autorange=False
    ),
    yaxis=dict(
        title='Product Name',
        automargin=True
    ),
    height=800,
    width=1500,
    margin=dict(r=20, t=50, l=150, b=50)
)
fig_max_discount.show()

print(max_discount_by_product_name*100)
```

##Sentiment analysis

```python
from textblob import TextBlob
def get_sentiment(text):
    return TextBlob(str(text)).sentiment.polarity

#Applying sentiment analysis
df['review_sentiment'] = df['review_content'].apply(get_sentiment)
plt.figure(figsize=(18, 6))
sns.histplot(df['review_sentiment'], bins=20, kde=True, color='green')
plt.title('Sentiment Distribution of Reviews')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()
```
